SETS

```python
#**Set** Python ka ek data type hai jo multiple values ko store karta hai. Set ki sab se
#  important properties yeh hain ke yeh **unordered** hota hai, yani values ka koi fixed order
#  nahi hota. Set sirf **unique values** ko rakhta hai, is liye agar koi value ek se zyada baar 
# aaye to set uske **duplicate values automatically remove** kar deta hai. Isi wajah se set tab 
# useful hota hai jab hame repeated values ko hatana ho aur sirf unique data chahiye ho.
```

```python
#duplicate remove kar deta hai
first_set={1,2,34,5,1} # Fist method
first_set
```

```python
# 2nd Method 

second_set = set([1,2,34,5,1])
second_set
```

```python
# Set unordered hota hai.Index nahi hota.
second_set[0]
```

```python
#User se 2 names leta hai.
# Append karta hai.
# Baad mein print karta hai.
# Ye list banane ki practice hai.
names = []

for i in range(2): 
    name = input("Enter your name: " )
    names.append(name)
    print(names)

for i in names:
    print(i)
```

Nested loop

```python
nested_lists = [[1,2,4,5,6,89,9],[1,4,67,9,9,8,9,10,2,3],["Starc","Kohli","Root"]]

# elements in first list 3
for i in nested_lists:
    print(i)
    # 1st iteration [1,2,4,5,6,89,9]
    # 2nd iteration [1,4,67,9,9,8,9,10,2,3]
    # 3rd iteration ["Starc","Kohli","Root"]
    for j in i:
        print(j)
```
