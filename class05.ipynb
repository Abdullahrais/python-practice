```python
clean_cities=["New York","Los Angeles","Chicago","Houston"]
user_input=""
while user_input !="q":
    user_input =input("Enter your city name :")
    if user_input != "q": 
        for a_clean_city in clean_cities:
            if user_input == a_clean_city:
                print("it is a clean city")
    else:
       break
```
